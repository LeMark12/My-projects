# Messing around with data from lecture 4

Load some data and check the head to make sure it seems correct.

```python
myfile='movie_metadata.csv'
#standard imports for when working with pandas
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd

#read data from a csv file
df = pd.read_csv(myfile,index_col=None)

df.head()
```

Sort the mean imdb score and budget across films by genre. Set a minimum imdb score and filter according to that. We are interested only in high scores so we are only looking at films with good ratings. Take a look at what this returns

```python
genreseries=df.groupby('genres')['imdb_score','budget'].mean() # get mean imdb rating and budget by genre
min_imdb_score=8 # set some minimum value for imdb rating to just look at well-rated films
top_rated_genre = genreseries[genreseries['imdb_score']>min_imdb_score] # apply rating filter
top_rated_genre = top_rated_genre.sort_values('imdb_score', ascending=False)
top_rated_genre # show data frame
```

Good. There seems to be some well-rated genres that we could propose as better options to make a film. But which genres are in our budget? Filter with a max budget and sort by imdb score, take a look again.

```python
max_budget = 1500000 # set some budget to filter by
topgenreseries=df.groupby('genres')['imdb_score','budget','gross'].mean() # get mean imdb rating, budget and gross by genre

budget_genres = topgenreseries[topgenreseries['budget']<max_budget] # filter by budget
budget_genres = budget_genres.sort_values('imdb_score',ascending=False)
budget_genres # show dataframe
```

The Documentary|Family genre seems to be well-rated and within our budget. Let's take a look at the details more closely.

```python
df[df['genres']=='Documentary|Family'] # grab just documentary|family genre by name
```

Maybe not good evidence or a suitable genre after all. Let's take a look at another genre...

```python
df[df['genres']=='Biography|Documentary|Music'] # grab another genre by name
```

OK, this isn't going well. Time to think of something else...

```python
max_budget=1500000 # set some budget to filter by

potential_genres=df.loc[:,['genres','imdb_score','budget']].groupby('genres') # grab some imdb, budget and genre info and sort by genre
imdb_mean=potential_genres.mean().reset_index('genres')

new_budget_genre_series = imdb_mean[imdb_mean['budget']<max_budget] # filter by budget
new_budget_genre_series = new_budget_genre_series.sort_values('imdb_score',ascending=False)
new_budget_genres = new_budget_genre_series['genres']
# new_budget_genre_series

good_movies = df[df.genres.isin(new_budget_genres)] # get all data from original data frame, index by things only in new_budget_genre_series
good_movies
```
